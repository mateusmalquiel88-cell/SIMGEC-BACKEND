````python
# Carregar o arquivo Excel
data_path = os.path.join(os.getcwd(), '07 - Cubal  - Criação e Recriação de Escolas.xlsx')
print('Caminho do arquivo:', data_path)

excel_data = pd.read_excel(data_path, sheet_name=0, dtype=str)
print('Dimensões do DataFrame:', excel_data.shape)
print('Colunas do arquivo:')
print(excel_data.columns.tolist())

# Mostrar as primeiras linhas
excel_data.head(10)
````

```markdown
## Definir funções de processamento
Criar funções reutilizáveis para transformar, limpar e analisar os dados.
```

````python
# Funções de processamento
def normalize_column_names(columns):
    return [str(col).strip() for col in columns]


def count_non_empty_codes(df, code_column='CÓDIGO'):
    return df[code_column].notna().astype(int).sum()


def summarize_column_values(df, column, top_n=20):
    return df[column].value_counts(dropna=True).head(top_n)


def map_and_validate(df, mapping):
    mapped_df = df.rename(columns=mapping)
    mapped_df = mapped_df[[col for col in mapping.values() if col in mapped_df.columns]]
    mapped_df['codigo_valido'] = mapped_df['codigo'].notna() & (mapped_df['codigo'].str.strip() != '')
    mapped_df['nome_valido'] = mapped_df['nome'].notna() & (mapped_df['nome'].str.strip() != '')
    return mapped_df
````

```markdown
## Executar o fluxo de trabalho
Aplicar as funções ao DataFrame e inspecionar os resultados.
```

````python
# Normalizar nomes de colunas
excel_data.columns = normalize_column_names(excel_data.columns)

# Contar códigos e nomes válidos
codigo_count = count_non_empty_codes(excel_data, code_column='CÓDIGO')
print('Registros com CÓDIGO não vazio:', codigo_count)

nome_count = excel_data['DESIGNAÇÃO DA ESCOLA'].notna().astype(int).sum()
print('Registros com DESIGNAÇÃO DA ESCOLA não vazio:', nome_count)

# Resumo das primeiras 20 entradas de CÓDIGO
print('\nPrimeiros 20 valores de CÓDIGO:')
print(excel_data['CÓDIGO'].head(20).tolist())

# Verificar registros sem código mas com nome
missing_codes = excel_data[(excel_data['CÓDIGO'].isna() | (excel_data['CÓDIGO'].str.strip() == '')) & excel_data['DESIGNAÇÃO DA ESCOLA'].notna()]
print('\nRegistros sem CÓDIGO mas com nome (até 20):', len(missing_codes))
missing_codes[['DESIGNAÇÃO DA ESCOLA']].head(20)
````

```markdown
## Visualizar e salvar resultados
Gerar gráficos simples e exportar os resultados de análise.
```

````python
# Visualizar a distribuição dos códigos válidos
plt.figure(figsize=(12, 6))
plt.hist(excel_data['CÓDIGO'].dropna().astype(str), bins=20)
plt.title('Distribuição de CÓDIGO das escolas')
plt.xlabel('CÓDIGO')
plt.ylabel('Frequência')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Exportar resumo para CSV
summary = pd.DataFrame({
    'total_linhas': [excel_data.shape[0]],
    'codigo_nao_vazio': [codigo_count],
    'nome_nao_vazio': [nome_count],
    'linhas_sem_codigo_com_nome': [len(missing_codes)]
})
summary_path = os.path.join(os.getcwd(), 'excel_inspection_summary.csv')
summary.to_csv(summary_path, index=False)
print('Resumo exportado para:', summary_path)
````
